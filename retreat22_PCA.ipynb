```python
traindata = np.vstack([
    np.random.multivariate_normal([1,1],np.eye(2)*0.1,100),
    np.random.multivariate_normal([-1,-1],np.eye(2)*0.1,100),
    ])

testdata = np.vstack([
    np.random.multivariate_normal([1,1],np.eye(2)*1,100),
    np.random.multivariate_normal([-1,-1],np.eye(2)*1,100),
    ])

labels = np.hstack([np.zeros(100),np.ones(100)])
```

```python


def sigmoid(x):
    return 1.0/(1.0+np.exp(-x*20))
```

```python
b = 0 
W = np.random.normal([2,1]); 
alpha = 0.01
maxiteration = 10000

e = 1
e_list = [] #[maxiteration,1]);
counter = 0
while e > 0.0001: #or counter<=maxiteration-1:
    y = sigmoid(np.dot(traindata,W))
    W = W + alpha*np.dot((labels-y).T,traindata) 
    
    e = np.abs(labels-y).mean()
    e_list.append(e) #[counter,0] = e
    counter = counter + 1
    
```
